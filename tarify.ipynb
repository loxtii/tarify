```python
#        
model = DecisionTreeClassifier(random_state=FIXED_NUMBER)
```

```python
#     fit()
model.fit(features_train, target_train);
```

   accuracy -        /  

```python
#     ( )
train_predictions = model.predict(features_train)
train_accuracy = accuracy_score(target_train, train_predictions)
```

```python
#    
valid_predictions = model.predict(features_valid)
valid_accuracy = accuracy_score(target_valid, valid_predictions)
```

```python
# print(" accuracy   :", round(train_accuracy,2))
print(" accuracy   :", round(valid_accuracy, 2))
```

#####             

```python
best_model_tree = None
best_result_tree = 0
best_depth_tree = 1

#    
depth_tree = []
#   accuracy   
accuracy_train_tree = []
#   accuracy   
accuracy_valid_tree = []

#       1  15
for depth in range(1, 16):
    model_tree = DecisionTreeClassifier(random_state=FIXED_NUMBER, max_depth=depth)
    model_tree.fit(features_train, target_train)
    
    #      
    predictions_train = model_tree.predict(features_train)
    #      
    predictions_valid = model_tree.predict(features_valid) 

    accuracy_train = accuracy_score(target_train, predictions_train)
    accuracy_valid = accuracy_score(target_valid, predictions_valid)

    depth_tree.append(depth)
    accuracy_train_tree.append(accuracy_train)
    accuracy_valid_tree.append(accuracy_valid)

    if accuracy_valid > best_result_tree:
        best_model_tree = model_tree
        best_result_tree = accuracy_valid
        best_depth_tree = depth
```

```python
print('  " ":', best_model_tree)
print(' accuracy   " ":', round(best_result_tree, 3))
print('    " ":', best_depth_tree)
```
