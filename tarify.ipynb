### Разбивка данные на выборки

#### Разделим данные на общие признаки и целевой признак

Для нашей задачи выделим общие признаки для обучения и запишем их в переменную features и категориальный целевой признак is_ultra, который сохраним в переменную target.

```python
# ранее мы решили исключить minutes
features = df.drop(['minutes', 'is_ultra'], axis=1) 
target = df['is_ultra']
```

```python
# проверим их размеры
features.shape, target.shape
```

3214 строчки на месте, ничего не потеряли. 

#### Разделим исходные данные на обучающую, валидационную и тестовую выборки

Данные разобьем на три части: обучающую, валидационную и тестовую. Размеры обучающего, тестового и валидационного наборов определим в пропорции  3 : 1 : 1 (60% : 20% : 20%).

```python
# Выделим тестовую выборку в размере 20% из общего масиива данных
features_working, features_test, target_working, target_test = train_test_split(
    features, target, test_size=0.20, random_state=FIXED_NUMBER)
```

```python
# Из оставшейся части выделим валидационную выборку в размере 25%, что будет составлять 20% из общего масиива данных
features_train, features_valid, target_train, target_valid = train_test_split(
    features_working, target_working, test_size=0.25, random_state=FIXED_NUMBER)
```

```python
# Посмотрим размеры получившихся выборок
features_train.shape, features_valid.shape, features_test.shape
```

```python
# Посмотрим размеры целевых признаков:
target_train.shape, target_valid.shape, target_test.shape
```

Проверим корректность разбивки

```python
size = df.shape[0]
print(f'Обучающая     выборка составляет {features_train.shape[0]/size:.1%} от исходного датасета.')
print(f'Валидационная выборка составляет {features_valid.shape[0]/size:.1%} от исходного датасета.')
print(f'Тестовая      выборка составляет {features_test.shape[0]/size:.1%} от исходного датасета.')
```

Наличествует пропорция 60/20/20. Всё как мы любим.

Проведем расчет значения accuracy - отношения числа правильных ответов к размеру обучающей / валидационной выборок

##### Оценим различные значения глубины дерева на валидационной выборке и найдем из них лучшее

```python
best_model_tree = None
best_result_tree = 0
best_depth_tree = 1

# список с глубиной дерева
depth_tree = []
# список значений accuracy на обучающей выборке
accuracy_train_tree = []
# список значений accuracy на валидационной выборке
accuracy_valid_tree = []

# Построим модели с глубиной дерева от 1 до 15
for depth in range(1, 16):
    model_tree = DecisionTreeClassifier(random_state=FIXED_NUMBER, max_depth=depth)
    model_tree.fit(features_train, target_train)
    
    # расчет качества модели на обучающей выборке
    predictions_train = model_tree.predict(features_train)
    # расчет качества модели на валидационной выборке
    predictions_valid = model_tree.predict(features_valid) 

    accuracy_train = accuracy_score(target_train, predictions_train)
    accuracy_valid = accuracy_score(target_valid, predictions_valid)

    depth_tree.append(depth)
    accuracy_train_tree.append(accuracy_train)
    accuracy_valid_tree.append(accuracy_valid)

    if accuracy_valid > best_result_tree:
        best_model_tree = model_tree
        best_result_tree = accuracy_valid
        best_depth_tree = depth
```

```python
print('Наилучшая модель "дерево решений":', best_model_tree)
print('Значение accuracy лучшей модели "дерево решений":', round(best_result_tree, 3))
print('Глубина дерева лучшей модели "дерево решений":', best_depth_tree)
```
